```python
import os
import shutil
```

```python
file_path = "C:\\project\\Model-Zoo\\Dental\\detection\\landmark\\spine_data\data1"
target_path = 'C:\\project\\Model-Zoo\\Dental\\detection\\landmark\\spine_data\data1'
g = os.walk(file_path)
i = 0
for path, dir_list, file_list in g:
    for name in file_list:
        #print(name)
        target_full_path = os.path.join(target_path,name)
        source_full_path = os.path.join(path,name)
        print(target_full_path,source_full_path)
        shutil.copy(source_full_path,target_full_path)
        i = i + 1
print(i)
```

```python
file_path = "C:\\project\\Model-Zoo\\Dental\\detection\\landmark\\spine_data\\data1\\1"
target_path = 'C:\\project\\Model-Zoo\\Dental\\detection\\landmark\\spine_data\\data1\\1'
g = os.walk(file_path)
i = 0
for path, dir_list, file_list in g:
    print("path:",path)
    for dir in dir_list:
        f_dir = os.path.join(path,dir)
        filenames = os.listdir(f_dir)
        for f in filenames:
            if(f.endswith(".csv")):
                csv_file_name = dir + ".csv"
                #print("f:",f)
                #print("target:",os.path.join(target_path,csv_file_name))
                shutil.copy(os.path.join(f_dir,f),os.path.join(target_path,csv_file_name))
            elif(f.endswith(".nii.gz")):
                v_file_name = dir + ".nii.gz"
                #print("V:",v_file_name)
                shutil.copy(os.path.join(f_dir,f),os.path.join(target_path,v_file_name))
            
        #print(dir)
        i = i + 1
print(i)
```
